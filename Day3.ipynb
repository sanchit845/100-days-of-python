STRINGS

A string is a value enclosed between single (' ') or double (" ") or triple quotes for multi line string (""" """)

A String can be sliced to get a part of the string.
eg. - "Sanchit" has a length of 7.

eg. - s = "Sanchit"
      print (s[1:4])

      O/P - anc

here, 1 is the index start and 4 is index end.
The ending index will not be included

Strings are immutable - it means it cannot be changed.

```python
s = "Sanchit"
print(s[1:4])
print(s[5])
print(s[-3])
print(s[:4])
print(s[::4]) #skip 4 values   
```

Functions of a string:

1. len() - It returns the length of the string.
2. .endswith() - returns a boolean value of whether the string ends with the given value.
3. .startswith() - returns a boolean value of whether the string starts with the given value.
4. capitalize() - capitalizes the first letter of the string only.
5. .lower() - converts the string into lowercase
6. .title() - makes the first word of each word in the string capital.
7. .find() - returns the index of the first occurance of the word we need to find.
8. lstrip() - removes white spaces from the left side of the string
9. rstrip() - removes white spaces from the right of the string.

```python
s = "    hello world    "

print(len(s))                    # Length of the string
print(s.endswith("world    "))   # Checks if string ends with "world    "
print(s.startswith("    hello")) # Checks if string starts with "    hello"
print(s.capitalize())            # Capitalizes the first letter
print(s.lower())                 # Converts to lowercase
print(s.title())                 # Capitalizes the first letter of each word
print(s.find("world"))           # Finds the index of "world"
print(s.lstrip())                # Removes spaces from the left
print(s.rstrip())                # Removes spaces from the right
```

ESCAPE SEQUENCE CHARACTERS:
1. \n - new line
2. \t - tab space
3. \" \" - used to insert double quotes in the string

```python
a = "Sanchit is a good boy. \nHe plays football."
print(a)
b = "Sanchit is not a good \"boy\"."
print(b)
```
