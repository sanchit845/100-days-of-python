Operators in python are used to perform operations on variables and values. Python divides the operators in the following groups:

1. Arithmetic Opetators: +,-,*,/
2. Comparison Operators: ==, >=, <=, !=
3. Logical Operators: and, or, not
4. Assignment Operators: =, +=, -=

```python
#Arithmetic Operators

a = 10
b = 5

print(a+b) #Addition
print(a-b) #Subtraction
print(a*b) #Multiplication
print(a/b) #Division
```

```python
#Assignment Operators

a = 10
b = a
c = 5
c += 5
a -= 5

print(b) #Assign value of a to b
print(c) #Increment c by 5
print(a) #Decrement a by 5
```

```python
#Logical Operators

a = True
b = False

print(a and b) #Logical AND
print(a or b) #Logical OR  
print(not a) #Logical NOT
```

```python
#Comparison Operators

a = 10
b = 10.0

print( a == b) #Equal to
print( a != b) #Not equal to
print( a > b) #Greater than
print( a < b) #Less than   
print( a >= b) #Greater than or equal to
print( a <= b) #Less than or equal to  
```

Type Function () : This function is used to know the type of the variable like integer, float, string etc.

```python
a = "String"
b = float(10)
c = int(10.5)

print(type(a)) #String
print(type(b)) #Float
print(type(c)) #Integer
```

Type casting means when we change the datatype of a variable given the condition that it is legally correct.

```python
a = 10
print(type(a))
```

```python
a=float(a)
print(a)
print(type(a))
```

So here we saw, how a was an integer first but by using float() function, we changed it's type to float
