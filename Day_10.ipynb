# Functions and Recursions

Functions is a group of statements performing a specific task.
A function can be reused by a coder to run a given set of instructions multiple times.

Syntax:

def func():


You have to call a function to run it. eg- func()

Q. Write a function to greet a person

```python
def greet(): #defining a function
    """Greet the user with their name."""
    return f"Hello, Sanchit!"

greet() #calling a function
```

## Types of Functions

### Built in Functions
- Already present in python ( eg- len, range, list)

### User defined function 
- Made by users for their code snippets ( eg- func(), greet())

## Parameters in a function

Parameters are passed while defining a function, variables that will be used on the program

The user can pass the values of the parameter while calling the function.

Q. Write a function to print square of a number entered by the user while calling the function.

```python
def num(number): #defining a function with a parameter
    """Return the square of the number."""
    return number ** 2

print(num(3)) #calling a function with a value
print(num(5)) #calling a function with a value
```

We can also specify the value when definfing the parameters:

def func(name = "Sanchit"):

this will take name as Sanchit until explicitly mentioned in the code.

# Recursion

Recursion is a function that calls itself.

Q. Write a program to write the factorial of a number entererd by the user.

```python
def factorial(n): #defining a function to calculate factorial
    """Return the factorial of a number."""
    if n==0 or n==1:
        return 1
    return n*factorial(n-1)

num=int(input("Enter a number to calculate its factorial: ")) #taking input from user
print(f"The factorial of {num} is {factorial(num)}") #calling the function and printing the result
```

```python
#Q1. Write a program using functions to find greatest of three numbers.

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))
def great(a,b,c):
    if a>b and a>c:
        print(a, "is the greatest number.")
    elif b>a and b>c:
        print(b, "is the greatest number.")
    else:
        print(c, "is the greatest number.")

great(a,b,c)
```

```python
#Q2. Write a Python program using function to convert Celsius to Fahrenheit.

def celsius_to_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

celsius = float(input("Enter temperature in Celsius: "))
fahrenheit = celsius_to_fahrenheit(celsius)
print(f"{celsius}°C is equal to {fahrenheit}°F")
```

```python
#vQ3. Write a recursive function to calculate the sum of first n natural numbers..

def sum_natural(n):
    if n<=0:
        return 0
    else:
        return n + sum_natural(n-1)
    
n=int(input("Enter a number to calculate the sum of first n natural numbers: "))
print(f"The sum of first {n} natural numbers is {sum_natural(n)}")
```

```python
''' Q4. Write a Python function to print first n lines of the following pattern:
***
**
*
'''

n = int(input("Enter the number of lines of pattern you want to print: "))
def pattern(n):
    for i in range(n,0,-1):
        print("*" * i)

pattern(n)
```

```python
#Q5. Write a Python function to convert centimeters to inches.

cm=int(input("Enter the number of centimeters to convert to inches: "))
def convert(cm):
    inches = cm / 2.54
    return inches

print(f"The value of {cm} cm in inches is {convert(cm)} inches.")
```

```python
#Q6. Write a Python function to remove a given word from a list and strip it of whitespace.

def remove_strip():
    list1 = ['  Sanchit  ', '  is  ', '  a  ', '  good  ', 'boy']
    stripped_list = [item.strip() for item in list1]
    rem = input("Enter the word you want to remove from the list: ")
    stripped_list.remove(rem)
    return stripped_list

remove_strip()
```

```python
#Q7. Write a Python function to print the multiplication table of a given number.

def print_multiplication_table(n):
    for i in range(1, 11):
        print(f"{n} x {i} = {n*i}")
        
num = int(input("Enter a number to print its multiplication table: "))
print_multiplication_table(num)
```
