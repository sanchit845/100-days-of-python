# LOOPS:

In python, a loop is a control structure that repeats a block of code multiple times until a specific condition is met or a defined number of repetitions is completed.

## Why Loops Are Used

Loops are useful because they help you:

- Avoid writing the same code again and again
- Process many items in a list, array, or string
- Repeat actions until a condition changes
- Automate repetitive tasks

# While Loop

this loop is executed when the given condition is true.

We define the initial value of the variable outside the loop and then we update the value inside the loop.

````python
## Example in Simple Words

## Instead of writing:

print("Hello")
print("Hello")
print("Hello")
print("Hello")
print("Hello")

print ("\nThis is with loop")

# you can use a loop:

for i in range(5):
    print("Hello")

# This repeats the same action 5 times.
````



## Types of Loops

### 1. For Loop
A `for` loop is used when you know how many times you want to repeat something.

```python
for i in range(5):
    print(i)
```

This prints:

```python
0
1
2
3
4
```

### 2. While Loop
A `while` loop keeps running as long as a condition is true.

```python
count = 0

while count < 5:
    print(count)
    count += 1
```

This also prints numbers from 0 to 4.

## Loop Control Statements

These statements help control the flow of a loop:

- `break` stops the loop immediately
- `continue` skips the current iteration and moves to the next one
- `pass` is a placeholder in Python (does nothing)

Example:

```python
for i in range(5):
    if i == 3:
        continue
    print(i)
```

This skips the number 3.

## Nested Loops

A nested loop is a loop inside another loop.

```python
for i in range(2):
    for j in range(3):
        print(i, j)
```

This creates combinations of values from both loops.

## Infinite Loop

If the condition of a loop never becomes false, the loop keeps running forever.

```python
while True:
    print("Hello")
```

This is called an infinite loop. It can crash the program if not stopped.


#### Q. Write a program to print the contents of a list using while loop

````python
L = [1,"Sanchit",3.5,True]
i=0
while i<len(L):
    print(L[i])
    i+=1
````

````python
# We can use else statement with for loop. The else statement is executed when the condition becomes false.

l=[1,3,8]
for i in l:
    print(i)
else:
    print("Loop completed successfully.")
````

````python
#Q1. Write a program to print the multiplication table of a given number using a for loop.

num = int(input("Enter a number: "))
for i in range(1, 11):
    print(num, "x", i, "=", num*i)
````

````python
#Q2. Write a program to greet all the person names stored in a list l and which starts with S.

# l = ["Harry", "Soham", "Sachin", "Rahul"]

l = ["Harry", "Soham", "Sachin", "Rahul"]
for name in l:
    if name.startswith("S"):
        print("Hello " + name)
````

````python
#Q3. Attempt problem 1 using a while loop.

num = int(input("Enter a number: "))
i = 1
while i<=10:
    print(num, "x", i, "=", num*i)
    i += 1
````

````python
#Q4. Write a program to find whether a given number is prime or not.

num = int(input("Enter a number: "))
for i in range(2, num):
    if num % i == 0:
        print(num, "is not a prime number")
        break
else:
    print(num, "is a prime number")
````

````python
#Q5. Write a program to find the sum of the first n natural numbers using a while loop.

n = int(input("Enter a number: "))
i = 1
sum = 0
while i <= n:
    sum += i
    i += 1
print("The sum of first", n, "natural numbers is:", sum)
````

````python
#Q6. Write a program to calculate the factorial of a given number using a for loop.

num = int(input("Enter a number: "))
factorual = 1
for i in range(1, num + 1):
    factorual *= i
print("The factorial of", num, "is:", factorual)
````
