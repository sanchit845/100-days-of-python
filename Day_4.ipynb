Lists:

A collection of similar or non similar elements with similar or different dataypes enclosed between [].
They are mutable - that menas the values of list can be changed.

The List is indexed as a string only.

We can add an element to the list, We can delete an element of the list, We can slice the string as per our wish.

Slice: list([starting:ending]) -> Ending value is not enclosed in the slicing.

```python
names = ["Sanchit","Aditya","Sagar","Prateek","Sahil"]
print(type(names))
```

```python
print(names[2]) 
print(names[-3])
```

```python
print(names[0:4]) #Slicing 
```

When we update a string : We get a new string in return as strings are Immutable.

But when we make a change in a list : We get the list updated as list is mutable, that is, it can be changed.

```python
#Functions of lists

names.append("Siddharth") #Adding an element to the list
print("Appended list is: " + str(names))
```

```python
#Functions of lists

names.reverse() #Reversing the list
print("Reversed list is: " + str(names))
```

Here we can see that our original list was names = ["Sanchit","Aditya","Sagar","Prateek","Sahil"] : but when we appended it, it made the changes in the original list.

```python
#Functions of lists

names.sort() #Sorting the list
print("Sorted list is: " + str(names))
```

```python
#Functions of lists

names.insert(2,"Rohit") #Inserting an element at a specific index
print("List after inserting an element is: " + str(names))
```

```python
#Functions of lists

names.pop() #Removing the last element from the list 
print("List after removing the last element is: " + str(names))
```

```python
#Functions of lists

names.remove("Rohit") #Removing a specific element from the list
print("List after removing a specific element is: " + str(names))
```
