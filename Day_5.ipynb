Tuple:

A tuple is an ordered collection of items. It's similar to a list, but in Python, a tuple is immutable, meaning you cannot change its contents after it is created.

We can access the elements of a tuple by traversing it, but we cannot modify the elements of a tuple.

```python
a = (1, "Sanchit",False, "Hello")
print(type(a)) #This will print the type of a which is tuple
```

```python
#Functions of a tuple

no = a.count(1) #This will count the number of times 45 is present in the tuple
print(no)
```

```python
#Functions of a tuple
print(a)
print(a.index("Hello")) #This will return the index of the first occurrence of "Sanchit" in the tuple
```

```python
#Functions of a tuple

print(len(a)) #gives the length of the tuple
```

```python
#Functions of a tuple

b = (10, 20, 30, 40)
print(max(b)) #gives the maximum value in the tuple
print(min(b)) #gives the minimum value in the tuple
```

```python
#Functions of a tuple

print(sorted(b)) #gives the sorted tuple
```

```python
#Functions of a tuple

print(sum(b)) #gives the sum of the tuple
```
