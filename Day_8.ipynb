CONDITIONAL EXPRESSION

It is used when we want an action performed based on a specific condition/conditons.

if - If the condition of the if part is true, the "if" block will get executed.

else - when "if" condition fails, the kernel jumps to the else and executes the else block.

elif - Used between if and else for multiple conditions

```python
age = int(input("Please enter your age: "))

if age < 18:
    print("You are a minor.")
elif age >= 18 and age < 65:
    print("You are an adult.")
else :
    print("You are a senior citizen.")

print("End of Program.")
```

This is called a If-Else ladder.

If any of the condition of if-elif-else becomes true, the other statements gets ignored and the control of the program jumps to the next line outside the if-elif-else ladder.

#1. Write a program to find the greatest of four numbers entered by the user.

```python
a = int(input("Enter a number: "))
b = int(input("Enter second number: "))
c = int(input("Enter a third number: "))
d = int(input("Enter a fourth number: "))
if a > b and a > c and a> d:
    print("The largest number is:", a)
elif b>a and b>c and b>d:
    print("The largest number is:", b)
elif c>a and c>b and c>d:
    print("The largest number is:", c)
else:
    print("The largest number is:", d)
```

#2. Write a program to find out whether a student has passed or failed if it requires a total of 40% and at least 33% in each subject to pass. Assume 3 subjects and take marks as input from the user.

#5. Write a program which finds out whether a given name is present in a list or not.

```python
names  = ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Heidi", "Ivan", "Judy"]
n = input("Enter the name you want to check: ").lower()
if (n in [name.lower() for name in names]):
    print("The name is present in the list.")
else:
    print("The name is not present in the list.")

print(n)
```
