Dictionaries

A collection of Key-Value pairs which are mutable and can be traversed as per need.

```python
marks = {"Sanchit" : 95,
         "Milan" : 85,
         "Mihir" : 75}
print(marks,"\n")
print(type(marks))
```

To find the record or a particular key, we can traverse it as per our needs - [O(1) time complexity]

```python
print(marks["Sanchit"])
```

Properties of dictionaries:

1. It is unordered
2. It is mutable
3. It is indexed
4. Cannot contain duplicate keys

Dictionary Methods:

1. items() -> returns a list of (key,value) tuples.
2. keys() -> returns all the keys.
3. update() -> Updates the dictionary with the supplied key-value pairs.
4. get() -> returns the value of the specified keys.

```python
#1.

print(marks.items())
```

```python
#2.

print(marks.keys())
```

```python
#3.

marks.update({"Sanchit" : 100})
print(marks)
```

```python
marks.get("Milan")
```

```python
print(marks)
```
