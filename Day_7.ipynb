Union and intersectiom:

Union means combining the unique values of 2 sets and writing them as a new set.

Intersection means, the unique values of 2 sets that are common in both sets.

```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
print("Union: ",s1.union(s2))
print("Intersection: ",s1.intersection(s2))
print(s1-s2)
```
